## 1. Compare the average weight of super powers who have Super Speed to those who do not.

```python
sup_speed = merged['Super Speed'] == True
```

```python
sup_slow = merged['Super Speed'] == False
```

```python
df_speed = merged.loc[sup_speed]
speed_weight = df_speed[['Weight']].mean()
speed_weight
```

```python
df_slow = merged.loc[sup_slow]
slow_weight = df_slow[['Weight']].mean()
slow_weight
```

```python
print (f'Average speedster ', speed_weight);
print (f"Average non-speedster ", slow_weight);
```
